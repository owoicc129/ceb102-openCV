# Simple plot 畫線

```python
#第一行 是用ju特別要寫的
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256, endpoint=True) #在負pi~正pi間產生256個點
C,S = np.cos(X), np.sin(X) #把256個點丟到cos sin裡

plt.plot(X,C) #畫線 (幾個點,對應的cos線)
plt.plot(X,S)
plt.title('this is title',fontsize=20)  #字體大小
plt.ylabel('y label') #標籤
plt.xlabel('x label')
plt.grid()  #畫背景格子


plt.savefig("./simple_plot.png",dpi=72) #存檔 但不能接在plt.show()後面用 不然會存到空白
plt.show() 
```

# Figure 獨立一張圖

# x/y limit圖邊框區間 & x/y tick 坐標軸顯示

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)


plt.figure()
# limits 圖邊框 顯示的區間
plt.xlim(X.min()*1.1, X.max()) #發現會黏到邊
plt.ylim(C.min()*1.1, C.max())

plt.plot(X,C)
plt.plot(X,S, color='red', linewidth=10)

#---------上下圖比較--------

plt.figure()
plt.xlim(X.min()*1.1, X.max()*1.1) #最小 最大 都 *1.1 就不黏邊
plt.ylim(C.min()*1.1, C.max()*1.1)

plt.plot(X,C)
plt.plot(X,S, color='red', linewidth=10)

plt.show()
```

# Line style 線的樣式 (有更多種可以google)

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

color = 'cornflowerblue'

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C = np.cos(X)

linestyles = ['-', '--', '-.', ':'] #不同線的形式 最底下實心 再來虛線 ...

for x, linestyle in enumerate(linestyles):
    plt.plot(X, C+x, linestyle=linestyle, color=color, linewidth=3)
    #       線  讓線不同高度                                粗度
    
plt.show()
```

# Marker 點的樣式
google matplotlib maker 可有更多樣式

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

color = 'cornflowerblue'

X = np.linspace(-np.pi, np.pi, 12, endpoint=True)
C = np.cos(X)

markers = ['o', 'v', '^', '<']

for x, linestyle in enumerate(markers):
    plt.plot(X, C+x, marker=markers[x], color=color, linewidth=1)
    #       線  讓線不同高度                                粗度
plt.show()
```

#### 左右型 長條圖

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

men_means = (20, 35, 30, 35, 27)  
women_means = (25, 32, 34, 20, 25) 

N = 5
ind = np.arange(N) #0~5 5不算
width = 0.35   #寬度   

#        x       y      寬度 
plt.bar(ind, men_means, width, label='Men') #藍

#      原本的x位置+寬度    y
plt.bar(ind + width, women_means, width, label='Women')  #橘

plt.ylabel('Scores') #y標籤
plt.title('Scores by group and gender')

#           刻度設在中間
plt.xticks(ind + width /2 , ('G1', 'G2', 'G3', 'G4', 'G5')) #x座標刻度
plt.legend(loc='best') #標籤介紹
plt.show()
```

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

countries = ['USA', 'GB', 'China', 'Russia', 'Germany'] # x
bronzes = np.array([38, 17, 26, 19, 15]) # y
silvers = np.array([37, 23, 18, 18, 10]) # y
golds = np.array([46, 27, 26, 19, 17]) # y

ind = [x for x, _ in enumerate(countries)] #取0~4 array

#        x     y       寬度       標籤            顏色          底下疊著誰
plt.bar(ind, golds, width=0.8, label='golds', color='gold', bottom=silvers+bronzes)
plt.bar(ind, silvers, width=0.8, label='silvers', color='silver', bottom=bronzes)
plt.bar(ind, bronzes, width=0.8, label='bronzes', color='#CD853F')

plt.xticks(ind, countries)  # x刻度(刻在哪,名稱)
plt.ylabel("Medals")
plt.xlabel("Countries")
plt.legend(loc="upper right") #標籤介紹
plt.title("2012 Olympics Top Scorers")

plt.show()
```

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
 
# Create data
N = 60
                                    #np.random.rand產生隨機樣本位於0~1中
g1 = (0.6 + 0.6 * np.random.rand(N), np.random.rand(N))  #(x,y)
g2 = (0.4+0.3 * np.random.rand(N), 0.5*np.random.rand(N))
g3 = (0.3*np.random.rand(N),0.3*np.random.rand(N))
 
data = (g1, g2, g3)
colors = ("red", "green", "blue")
groups = ("coffee", "tea", "water") 

for data, color, group in zip(data, colors, groups):
    x, y = data
    plt.scatter(x, y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)
               #      點的透明度             點的外框顏色     點大小  簡介標籤


plt.legend(loc=2) #2是左上 ,1是右上, 3是左下 ...
plt.title('Matplot scatter plot')
plt.show()
```

# Pie chart

```python
import matplotlib.pyplot as plt
 
labels = ['Cookies', 'Jellybean', 'Milkshake', 'Cheesecake']
sizes = [38.4, 40.6, 20.7, 10.3]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
                                                #陰影      #從哪個角度往左開始畫
    
plt.legend(patches, labels, loc="best")
plt.axis('equal') #一定要加 圓餅圖才不會怪怪
plt.tight_layout()  #不要互相壓到
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
# ref: https://matplotlib.org/examples/pie_and_polar_charts/pie_demo_features.html
# Pie chart, where the slices will be ordered and plotted counter-clockwise:

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # 爆炸程度 

fig1, ax1 = plt.subplots()
                                               #autopct顯示百分比
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')  

plt.show()
```

# Contours 等高線地形圖

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def f(x,y):
    # 高度
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n) #(幾到幾區間,畫幾個點)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y) #meshgrid把傳入的256*256座標點畫出來

#畫色塊             高度  level 透明度
plt.contourf(X, Y, f(X, Y),8, alpha=.75, cmap=plt.cm.hot)

# 畫線
C = plt.contour(X, Y, f(X, Y), 8, colors='black')
# adding label
plt.clabel(C, inline=True, fontsize=10)

plt.xticks(())
plt.yticks(())
plt.show()
```

# 3D data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

# X, Y value
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X ** 2 + Y ** 2)

# height value
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
"""
============= ================================================
        Argument      Description
        ============= ================================================
        *X*, *Y*, *Z* Data values as 2D arrays
        *rstride*     Array row stride (step size), defaults to 10
        *cstride*     Array column stride (step size), defaults to 10
        *color*       Color of the surface patches
        *cmap*        A colormap for the surface patches.
        *facecolors*  Face colors for the individual patches
        *norm*        An instance of Normalize to map values to colors
        *vmin*        Minimum value to map
        *vmax*        Maximum value to map
        *shade*       Whether to shade the facecolors
        ============= ================================================
"""

# I think this is different from plt12_contours
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.get_cmap('rainbow'))
"""
==========  ================================================
        Argument    Description
        ==========  ================================================
        *X*, *Y*,   Data values as numpy.arrays
        *Z*
        *zdir*      The direction to use: x, y or z (default)
        *offset*    If specified plot a projection of the filled contour
                    on this position in plane normal to zdir
        ==========  ================================================
"""

ax.set_zlim(-2, 2)

plt.show()
```
